```python
# Importamos la libreria OpenCv
import cv2 as cv
import PIL
# Importamos Pytesseract
import pytesseract
from pytesseract import Output
```

```python
myconfig = r"--psm 12 --oem 3"
#text = pytesseract.image_to_string(PIL.Image.open("pene.png"), config=myconfig)
#print(text)
```

```python
img = cv.imread('navas2.png')
img = cv.imread('navas.jpeg')
img = cv.imread('vinicus.png')
height, width, _ = img.shape

data = pytesseract.image_to_data(img,config=myconfig, output_type=Output.DICT)

#boxes = pytesseract.image_to_boxes(img, config=myconfig)
amount_boxes = len(data['text'])
for i in range(amount_boxes):
  if float(data['conf'][i]) >70:
    (x, y, width, height) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
    img = cv.rectangle(img, (x,y), (x+width, y+height), (0,255,0),2)
    img = cv.putText(img, data['text'][i], (x,y+height+20), cv.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2, cv.LINE_AA)
    print(data['text'][i])

cv2_imshow(img)
cv.waitKey(0)
cv.destroyAllWindows()
```
