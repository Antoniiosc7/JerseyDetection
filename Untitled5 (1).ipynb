```python
import re
import cv2
import pytesseract
from pytesseract import Output
from google.colab.patches import cv2_imshow

img = cv2.imread('navas2.png')

# Obtener datos de texto con coordenadas
d = pytesseract.image_to_data(img, output_type=Output.DICT)
n_boxes = len(d['text'])

# Patrón de fecha y letras
date_and_letters_pattern = '^[A-Za-z]+$'

# Iterar sobre las cajas de texto
for i in range(n_boxes):
    if int(d['conf'][i]) > 60:
        # Verificar si el texto coincide con el patrón
        if re.match(date_and_letters_pattern, d['text'][i]):
            # Obtener las coordenadas de la caja
            (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
            # Dibujar un rectángulo alrededor del texto
            img = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Mostrar la imagen con rectángulos dibujados
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
