```python
import cv2
import easyocr
from matplotlib import pyplot as plt

def detectar_texto(imagen):
    img = cv2.imread(imagen)

    # Instanciar EasyOCR y detectar texto
    reader = easyocr.Reader(['en'])
    resultado = reader.readtext(img)

    # Dibujar rectángulos alrededor del texto detectado
    for detection in resultado:
        box = detection[0]
        text = detection[1]
        score = detection[2]
        x1, y1 = int(box[0][0]), int(box[0][1])
        x2, y2 = int(box[1][0]), int(box[1][1])
        x3, y3 = int(box[2][0]), int(box[2][1])
        x4, y4 = int(box[3][0]), int(box[3][1])

        pts = [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]
        pts = [[[x1, y1], [x2, y2], [x3, y3], [x4, y4]]]
        pts = [pts]

        cv2.polylines(img, pts, isClosed=True, color=(255, 0, 255), thickness=2)
        cv2.putText(img, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 255), 2)

    # Mostrar la imagen con los rectángulos y texto detectados
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # Extraer y retornar el texto de la imagen
    texto = [detection[1] for detection in resultado]
    return ' '.join(texto)

texto = detectar_texto('navas.jpg')
print("El texto de la imagen es:", texto)
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt

# Instancia de EasyOCR
reader = easyocr.Reader(['en'])  # Especifica el idioma que se va a utilizar

# Carga de la imagen con múltiples camisetas (reemplaza 'ruta_imagen.jpg' por la ruta de tu imagen)
image_path = 'M.jpg'
img = cv2.imread(image_path)

# Carga el modelo pre-entrenado para la detección de objetos (en este caso, se utiliza el modelo pre-entrenado de Haar Cascade para camisetas)
cascade_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convierte la imagen a escala de grises para la detección
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detecta las camisetas en la imagen usando el modelo pre-entrenado de Haar Cascade
detections = cascade_classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Procesa cada detección para extraer el texto con EasyOCR
for (x, y, w, h) in detections:
    # Extrae la región de la imagen donde se encuentra la camiseta
    roi = img[y:y+h, x:x+w]

    # Utiliza EasyOCR para obtener el texto de la región de la camiseta
    result = reader.readtext(roi)

    # Filtra el texto para obtener el nombre y el número
    player_name = ""
    player_number = ""

    for detection in result:
        text = detection[1]
        if text.isalpha():
            player_name += text + " "
        elif text.isdigit():
            player_number = text

    # Muestra los resultados
    print("Nombre del jugador:", player_name)
    print("Número de la camiseta:", player_number)

    # Dibuja un rectángulo alrededor de la camiseta detectada
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Muestra la imagen con las camisetas detectadas y sus nombres/números
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
